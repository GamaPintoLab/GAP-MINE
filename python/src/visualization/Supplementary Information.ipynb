```python
complete_network = pd.DataFrame(columns = ['metric', 'method', 'clf', 'f_measure', 'f_measure_min', 'f_measure_max', 'precision', 'precision_min', 'precision_max', 'recall', 'recall_min', 'recall_max'])
complete_network[['metric', 'method', 'clf']] = df_plots_merge[(df_plots_merge['quality_metric'].isin(['f_measure'])) ].groupby(['metric', 'method', 'clf']).median().reset_index()[['metric', 'method', 'clf']]

complete_network[['f_measure_min', 'f_measure', 'f_measure_max']] = df_plots_merge[(df_plots_merge['quality_metric'].isin(['f_measure']))].groupby(['metric', 'method', 'clf']).quantile([0.25, 0.5,0.75]).round(3).unstack().values
complete_network[['precision_min', 'precision', 'precision_max']] = df_plots_merge[(df_plots_merge['quality_metric'].isin(['precision']))].groupby(['metric', 'method', 'clf']).quantile([0.25, 0.5,0.75]).round(3).unstack().values
complete_network[['recall_min', 'recall', 'recall_max']] = df_plots_merge[(df_plots_merge['quality_metric'].isin(['recall']))].groupby(['metric', 'method', 'clf']).quantile([0.25, 0.5,0.75]).round(3).unstack().values

complete_network['f_measure'] = complete_network.apply(lambda row: f"{row['f_measure']} ({row['f_measure_min']}-{row['f_measure_max']})", axis=1)
complete_network['precision'] = complete_network.apply(lambda row: f"{row['precision']} ({row['precision_min']}-{row['precision_max']})", axis=1)
complete_network['recall'] = complete_network.apply(lambda row: f"{row['recall']} ({row['recall_min']}-{row['recall_max']})", axis=1)

complete_network[['metric', 'method', 'clf', 'f_measure', 'precision', 'recall']].to_csv('../../reports/complete_network_table.csv', index=False)
complete_network[['metric', 'method', 'clf', 'f_measure', 'precision', 'recall']]
```
